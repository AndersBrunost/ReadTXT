```python
import datetime
import matplotlib.pyplot as plt

# Define the file path
file_path = 'Temperature.txt'

# Initialize empty lists to store data
dates_and_times = []
temperatures = []

# Read data from the file
try:
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) == 3:
                date_time_str, time_str, temp_str = parts
                date_time_str = f"{date_time_str} {time_str}"
                date_time = datetime.datetime.strptime(date_time_str, "%Y-%m-%d %H:%M:%S")
                temperature = float(temp_str)
                dates_and_times.append(date_time)
                temperatures.append(temperature)
except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
    exit(1)
except (ValueError, IndexError):
    print(f"Error: Invalid data format in '{file_path}'.")
    exit(1)

# Create a plot
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
plt.plot(dates_and_times, temperatures, marker='o', linestyle='-', markersize=5)
plt.title('Temperature Measurements Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (Celsius)')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
